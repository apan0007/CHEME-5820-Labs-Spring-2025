```julia
function world(a::Int64)::Float64

    # initialize -
    model = Dict{Int, Beta}();
    model[1] = Beta(2.0,2.0); # action 1
    model[2] = Beta(10.0,2.0); # action 2
    model[3] = Beta(2.0,10.0); # action 3

    # return -
    return model[a] |> d-> rand(d);
end;
```
